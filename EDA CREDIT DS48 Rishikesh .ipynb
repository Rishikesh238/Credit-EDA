```python
# columns greater than 40% can be dropped from application_data
app_dropped = (application_data.isnull().sum() / len(application_data))
app_dropped = list(app_dropped [app_dropped .values >= 0.40].index)
new_application =application_data.drop(labels=app_dropped , axis=1, inplace=False)
print('-'*100)
print('Shape of the data before deleting', application_data.shape)
print('-'*100)
print('Shape of the data after deleting', new_application.shape)
print('-'*100)
```

```python
# columns greater than 40% can be dropped from previous_application
prev_app_dropped = (previous_application.isnull().sum() / len(previous_application))
prev_app_dropped = list(prev_app_dropped[prev_app_dropped.values >= 0.40].index)
new_prev_application =previous_application.drop(labels=prev_app_dropped, axis=1, inplace=False)
print('-'*100)
print('Shape of the data before dropping', previous_application.shape)
print('-'*100)
print('Shape of the data after dropping', new_prev_application.shape)
print('-'*100)
```

# Categorical Univariate Analysis

```python
# Creating a reusable code for ploting
def univariate_cat(df,col,title,hue ='TARGET'):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='magma') 
    ax.set_title(title,fontdict={'fontsize' : 20, 'fontweight' : 3, 'color' : 'Blue'}) 
    ax.legend(['Repayer','Defaulter'])    
    plt.show()
```

### Top 10 correlation for the Repayers dataframe

```python
corr_repayer = Repayer_df.corr()
corr_df_repayer = corr_repayer.where(np.triu(np.ones(corr_repayer.shape),k=1).astype(np.bool)).unstack().reset_index()
corr_df_repayer.columns =['VAR1','VAR2','Correlation']
corr_df_repayer.dropna(subset = ["Correlation"], inplace = True)
corr_df_repayer["Correlation"]=corr_df_repayer["Correlation"].abs() 
corr_df_repayer.sort_values(by='Correlation', ascending=False, inplace=True) 
corr_df_repayer.head(10)
```
